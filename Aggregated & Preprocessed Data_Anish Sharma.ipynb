### -> Importing Required Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### 1). Loading Raw Data as Dataframe.

```python
# Importing the Dataset as Dataframe

df = pd.read_csv('Dataset_Uber Traffic.csv')

#Renaming a Column 
df = df.rename(columns = {'Vehicles':'Vehicle_Counts'})

df.head(50)
```

### 2). Data Cleaning.

#### (i). Handeling Missing Values using Imputation.

```python
#  Mean, Median, and Mode Imputation 
mean_imputation = df['Vehicle_Counts'].fillna(df['Vehicle_Counts'].mean()) 
median_imputation = df['Vehicle_Counts'].fillna(df['Vehicle_Counts'].median()) 
mode_imputation = df['Vehicle_Counts'].fillna(df['Vehicle_Counts'].mode().iloc[0]) 
  
print("\nImputation using Mean:") 
print(mean_imputation.head(10))
  
print("\nImputation using Median:") 
print(median_imputation)
  
print("\nImputation using Mode:") 
print(mode_imputation.tail(10))
```

#### (ii). Handeling Missing Values using Removal.

```python
# Removing rows with missing values 
df_cleaned = df.dropna() 
  
print("\nDataFrame after removing rows with missing values:") 
print(df_cleaned.head(20))

# Check for any remaining missing values
print("\nMissing Values Check:")
print(df_cleaned .isnull().sum())
```

#### (iii). Removing Duplicates.

```python
# Identify duplicate records
duplicate_mask = df.duplicated()
duplicates = df[duplicate_mask]
print(duplicates)

# Remove duplicates
data = df.drop_duplicates()
```

#### (iv). Correcting Datatypes.

```python
# Convert timestamp to datetime
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Convert junction to integer
df['Junction'] = df['Junction'].astype(int)

# Convert vehicle_counts to integer (assuming it should be)
df['Vehicle_Counts'] = df['Vehicle_Counts'].astype(int)

print(df.dtypes)
```

### 3). Aggregating Traffic Data.

#### Compiling Traffic Data into Hourly Intervals for each Junction.

```python
# NOTE :- 0 denotes the Missing Values

df['DateTime'] = pd.to_datetime(df['DateTime'])
df['Hour'] = df['DateTime'].dt.hour
df['Date'] = df['DateTime'].dt.date

hourly_traffic = df.groupby(['Junction', 'Date', 'Hour','DateTime']).agg({'Vehicle_Counts': 'sum'}).reset_index()

# Display aggregated hourly traffic data
print("\nHourly Traffic Data for each Junction:")
print(hourly_traffic.head(50))
```

### 4). Preprocessing the Data.

#### Normalizing the Data to Facilitate Comparison across different time Periods and Junctions.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
hourly_traffic['vehicle_counts_scaled'] = scaler.fit_transform(hourly_traffic[['Vehicle_Counts']])

# Display normalized data
print("\nNormalized Hourly Traffic Data:")
print(hourly_traffic.head(10))
```

### 5). Creating New Features from Raw Data.
#### (i). Generating Time-based Features such as Hour of the Day, Day of the Week, and Month.

```python
hourly_traffic['Hour_of_Day'] = hourly_traffic['DateTime'].dt.hour
hourly_traffic['Day_of_Week'] = hourly_traffic['DateTime'].dt.dayofweek
hourly_traffic['Month'] = hourly_traffic['DateTime'].dt.month

print("\nHourly Traffic Data with Time-based Features:")
print(hourly_traffic.head(10))
```

#### (ii). Developing Lag Features by including Traffic Data from previous Hours or Days to capture Temporal Dependencies.

```python
hourly_traffic['Lag_Vehicle_Counts'] = hourly_traffic.groupby(['Junction'])['Vehicle_Counts'].shift(1)

print("\nHourly Traffic Data with Lag Features:")
print(hourly_traffic.head(10))
```

#### (iii). Creating Binary Indicators for Weekends and Special Events to Account for their Impact on Traffic.

```python
# Binary Indicator for Weekends
hourly_traffic['Weekend'] = hourly_traffic['DateTime'].dt.dayofweek // 5

# Binary Indicator for Special Events
hourly_traffic['Special Events'] = hourly_traffic['DateTime'].dt.dayofweek // 5

print("\nHourly Traffic Data with Indicators for Weekends & Special Events:")
print(hourly_traffic.head(10))
```

### 6). Evaluating Feature Importance.

#### Using Statistical Methods and Machine learning Techniques (e.g., Correlation Analysis, Feature Importance from Tree-based Models) to assess the Relevance of each Feature.

```python
from sklearn.ensemble import RandomForestRegressor


X = hourly_traffic[['Hour_of_Day', 'Day_of_Week', 'Month', 'Lag_Vehicle_Counts', 'Weekend', 'Special Events']].fillna(0)
y = hourly_traffic['Vehicle_Counts']

# Initialize Random Forest Regressor
model = RandomForestRegressor()
model.fit(X, y)

# Get feature importances
feature_importances = model.feature_importances_

# Map feature names to their importance scores
feature_names = X.columns
feature_importance_dict = dict(zip(feature_names, feature_importances))

# Sort features by importance
sorted_features = sorted(feature_importance_dict.items(), key=lambda x: x[1], reverse=True)

# Print feature importances
print("\nFeature Importance:")
for feature, importance in sorted_features:
    print(f"{feature}: {importance}")

# Compute correlation matrix
correlation_matrix = hourly_traffic[['Vehicle_Counts', 'Hour_of_Day', 'Day_of_Week', 'Month', 'Lag_Vehicle_Counts', 'Weekend','Special Events']].corr()

# Plotting correlation matrix
plt.figure(figsize=(8, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
